```python
class Solution(object):
    def longestPalindrome(self, s):
        current_longest = -1 # currently longest string
        pivot = 0 # middle point
        if_middle = True
        
        for word in range(len(s)):
            try:
                if s[word] == s[word + 1]: #if consequtive repetition
                    counter = 1 #counter for pairs
                    if word - counter < -1:
                        continue
                    else:
                        try:
                            while s[word - counter] == s[word + counter + 1] and word - counter >= 0:
                                print(counter)
                        except:
                            pass
                        length = counter
                        if length * 2 > current_longest:
                            current_longest = length
                            pivot = word
                            if_middle = False
            except:
                pass
            
            try:
                if s[word - 1] == s[word + 1]: #if one-spaced repetition
                    if word - 1 < 0:
                        continue
                    else:
                        counter = 0
                        try:
                            while s[word - counter - 1] == s[word + counter + 1] and word - counter - 1 >= 0:
                                counter = counter + 1
                        except:
                            pass
                        length = counter
                        if length * 2 + 1 > current_longest:
                            current_longest = length
                            pivot = word
                            if_middle = True
            except:
                pass
        if current_longest == -1:
            return s[0]
        elif if_middle: #if one-spaced repetition
            return s[pivot - current_longest : pivot + 1 + current_longest]
        else:
            return s[pivot - current_longest + 1 : pivot + 1 + current_longest]
```
